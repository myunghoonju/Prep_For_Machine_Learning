```
import numpy as np
```

```
#use numpy.linlg
# A = USV : S is a diagonal matrix which means most of its entries are zeros hence this is a sparse matrix
from numpy.linalg import svd
```

```
A = np.array([[4,0],[-3,5]])

U,S,V = svd(A)
```

```
print("Left singular \n {}".format(U))
```

```
print("Singular values \n {}".format(np.diag(S)))
```

```
print("Right singular \n {}".format(V))
```

```
# @ is matrix multiplication same as np.matmul

print(U @ np.diag(S) @ V)
```

```
from sklearn.decomposition import TruncatedSVD
```

```
A = np.array([[-1, 2, 0], [2, 0, -2], [0, -2, 1]])

print("Original Matrix: \n {}".format(A))
```

```
svd =  TruncatedSVD(n_components = 2)
A_transf = svd.fit_transform(A)
```

```
print("Singular values:")
print(svd.singular_values_)
```

```
print("Transformed Matrix after reducing to 2 features:")
print(A_transf)
```

```
#faster Truncated svd: randomized svd
from sklearn.utils.extmath import randomized_svd

A = np.array([[-1, 2, 0], [2, 0, -2], [0, -2, 1]])
u, s, vt = randomized_svd(A, n_components = 2)

print("Left Singular Vectors:")
print(u)

print("Singular Values:") 
print(np.diag(s))

print("Right Singular Vectors:") 
print(vt)
```
